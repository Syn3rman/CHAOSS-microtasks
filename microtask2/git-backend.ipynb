```python
import pprint
from perceval.backends.core.git import Git
```

```python
uri = "https://github.com/Syn3rman/CHAOSS-microtasks"
repo_path = "./microtasks"
git = Git(uri=uri, gitpath=repo_path)
```

```python
# get all commits from a repo
for commit in git.fetch(category="commit"):
    pprint.pprint(commit["data"])
```

```python
commits = list(git.fetch(category="commit"))
# get total number of commits
num_commits = len(commits)
# get the latest 10 commits
latest_commits = commits[-10:]
```

```python
for commit in latest_commits:
    print("[Author]: {}, [Commit message]: {}\n".format(commit["data"]["Author"], commit["data"]["message"]))
    print("Files changed: ")
    print(*list(file["file"] for file in commit["data"]["files"]), "\n\n")
```
